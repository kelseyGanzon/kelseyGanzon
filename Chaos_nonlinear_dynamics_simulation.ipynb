```python
import numpy as np
import matplotlib.pyplot as plt

def map(x, alpha):
    return alpha*np.sin(np.pi*x)

def bifurcation_diagram(alpha_vals, x0, n_skip, n_iter):
    """
    Generates a bifurcation diagram for the map equation x_{n+1} = alpha*sin(pi*x_n)
    """
    # Set up storage for our results
    results = []

    # Iterate through each value of alpha
    for alpha in alpha_vals:
        # Set the initial condition
        x = x0

        # Iterate to allow transients to decay
        for i in range(n_skip):
            x = map(x, alpha)

        # Iterate and store the results
        for i in range(n_iter):
            x = map(x, alpha)
            results.append([alpha, x])

    # Return the results as a numpy array
    return np.array(results)
# Set up the parameters for our bifurcation diagram
alpha_vals = np.linspace(0, 1, 1000)
x0 = 0.1
n_skip = 100
n_iter = 1000

# Generate the bifurcation diagram
results = bifurcation_diagram(alpha_vals, x0, n_skip, n_iter)

# Plot the results
plt.plot(results[:,0], results[:,1], '.', markersize=0.1)
plt.xlabel('alpha')
plt.ylabel('x')
plt.show()
```
